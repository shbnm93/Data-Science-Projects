```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
import requests
import time
from bs4 import BeautifulSoup as bts
```

```python
def getAndParseURL(url):
    result = requests.get(url ,headers={"User-Agent":"Chrome/96.0.4664.92"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
PAGE_NUM = []
def find_page_num() :
  html = getAndParseURL("https://www.n11.com/telefon-ve-aksesuarlari/cep-telefonu")
  for i in range(1,51) :
            PAGE_NUM.append("https://www.n11.com/telefon-ve-aksesuarlari/cep-telefonu" + "?pg=" +str(i))
  return PAGE_NUM
find_page_num()
```

```python
ALL_PRODUCT = []
def all_product():
  for i in PAGE_NUM:
      html = getAndParseURL(i)
      for result in html.findAll("a",{"class":"plink"}):
        ALL_PRODUCT.append(result.get("href"))
  return ALL_PRODUCT
all_product()
```
