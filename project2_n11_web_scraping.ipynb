```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
import requests
import time
from bs4 import BeautifulSoup as bts
```

```python
def getAndParseURL(url):
    result = requests.get(url ,headers={"User-Agent":"Chrome/96.0.4664.92"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
PAGE_LINKS = []
def getPageLinks() :
    for i in range(1,51) :
            PAGE_LINKS.append("https://www.n11.com/telefon-ve-aksesuarlari/cep-telefonu" + "?pg=" +str(i))
    return PAGE_LINKS
getPageLinks()
```

```python
PRODUCT_LINKS = []
def getProductLinks():
  for pageLink in PAGE_LINKS:
      html = getAndParseURL(pageLink)
      for result in html.findAll("a",{"class":"plink"}):
        PRODUCT_LINKS.append(result.get("href"))
  return PRODUCT_LINKS
getProductLinks()
```
