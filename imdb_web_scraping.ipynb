```python
import requests
from bs4 import BeautifulSoup
Url = "https://www.imdb.com/chart/top/"
R = requests.get(Url)
print(R.text)
```

```python
Soup = BeautifulSoup(R.text, "html5lib")
print(Soup)
```

```python
List = Soup.find("tbody", {"class":"lister-list"}).find_all("tr")
for Film in List:
  print(Film)
```

```python
List = Soup.find("tbody", {"class":"lister-list"}).find_all("tr")
for Film in List:
  Name = Film("td",{"class":"titleColumn"}).a.text
  print(Name)
```
